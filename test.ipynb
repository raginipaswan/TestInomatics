```python
# Read the CSV files into DataFrames
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')

# Assuming 'movieId' and 'rating' columns are present in the ratings DataFrame
# If not, adjust the column names based on your actual data
popularity_df = ratings_df.groupby('movieId')['rating'].count().reset_index(name='ratings_count')

# Merge the popularity data with movie information
merged_df = pd.merge(popularity_df, movies_df, on='movieId', how='left')

# Filter Sci-Fi movies
sci_fi_movies = merged_df[merged_df['genres'].str.contains('Sci-Fi')]

# Sort Sci-Fi movies based on ratings count in descending order
sorted_sci_fi_movies = sci_fi_movies.sort_values(by='ratings_count', ascending=False)

# Get the third most popular Sci-Fi movie
third_most_popular_sci_fi_movie = sorted_sci_fi_movies.iloc[2]

# Display the result
print("Third Most Popular Sci-Fi Movie Based on Number of User Ratings:")
print(third_most_popular_sci_fi_movie[['title', 'ratings_count']])
```

```python
movies_df = pd.read_csv('movies.csv')
ratings_df = pd.read_csv('ratings.csv')
sci_fi_movies = movies_df[movies_df['genres'].str.contains('Sci-Fi')]

# Merge with ratings to get average ratings
average_ratings = pd.merge(ratings_df, sci_fi_movies, on='movieId', how='inner').groupby('movieId')['rating'].mean().reset_index(name='average_rating')

# Find the movieId with the highest average rating
highest_rated_sci_fi_movie = average_ratings.loc[average_ratings['average_rating'].idxmax()]

# Display the result
print("MovieId of the Sci-Fi movie with the highest IMDB rating:", highest_rated_sci_fi_movie['movieId'])
```
